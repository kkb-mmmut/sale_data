```python
# Import the Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.request
```

```python
# Load the dataset
df = pd.read_excel('SaleData.xlsx')
```

```python
df.to_csv('data.csv')
```

```python
# Print the dataset file save into data.csv
print(df)
```

```python
print(df.head())
```

```python
print(df.describe())
```

```python
#  Total Sales by Product Category
category_sales = df.groupby('Item')['Units'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=category_sales.index, y=category_sales.values, palette="viridis")
plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python

#  Monthly Sales Trend
df['Month'] = df['OrderDate'].dt.month
monthly_sales = df.groupby('Month')['Units'].sum()
plt.figure(figsize=(10, 5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o', color='b')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid()
plt.show()
```

```python
#  Managers Distribution
payment_method_counts = df['Manager'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=payment_method_counts.index, y=payment_method_counts.values, palette="pastel")
plt.title("Sale Distribution")
plt.xlabel("Manager Names")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
